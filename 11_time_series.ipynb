```
import numpy as np
import seaborn as sns
```

```
def create_time_series():
  freq = (np.random.random()*0.5) + 0.1
  ampl = np.random.random() + 0.5
  noise = [np.random.random()*0.3 for i in range(50)]
  x = np.sin(np.arange(0,50)*freq)*ampl+noise
  return x
#print(freq)
#'print(ampl)
#'print(noise)
#'print(x)
```

```
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
for i in range(0, 5):
  sns.tsplot( create_time_series(), color=flatui[i%len(flatui)] );  # 5 series
```
